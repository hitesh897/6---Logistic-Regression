```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
```

```python
# Importing Dataset
bank=pd.read_csv('bank-full.csv',sep=";")
```

```python
bank
```

### EDA

```python
bank.info()
```

```python
# Encoding of Categorical Values
data1=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
data1
```

```python
# All columns
pd.set_option("display.max.columns", None)
data1
```

```python
data1.info()
```

```python
# Binary encoding if bindary o/p variables
data1['default'] = np.where(data1['default'].str.contains("yes"),1,0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"),1,0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"),1,0)
data1['y'] = np.where(data1['y'].str.contains("yes"),1,0)
data1
```

```python
# Replacing Encoding for month categorical variable
data1['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
data1=data1.replace(order)
data1
```

```python
data1.info()
```

### Model Building

```python
# Diving our data into input and output variables
x=pd.concat([data1.iloc[:,0:11],data1.iloc[:,12:]],axis=1)
y=data1.iloc[:,11]
```

```python
# Logistic Regression Model
classifier=LogisticRegression()
classifier.fit(x,y)
```

### Model Predictions

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

### Testing Model Accuracy

```python
# Confusion Matrix
confusion_matrix=confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# Model Accuracy
(39107+1282)/(39107+815+4007+1282)
```

```python
# Accuracy is 0.8933 which is greater than 0.5 Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0
classifier.predict_proba(x)[:,1]
```

```python
# ROC Curve plotting
# Finding AIC Value
fpr,tpr,threshold=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area = %0.2f)' %auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy', auc)
```
